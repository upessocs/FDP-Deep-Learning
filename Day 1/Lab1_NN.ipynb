# Base Structure of training NN

Define Model or network

```
import torch
import torch.nn as nn
import torch.optim as optim


# Define the model
class MyModel(nn.Module):
  def __init__(self):
    super(MyModel, self).__init__()
    self.linear = nn.Linear(10, 10)
    self.linear1 = nn.Linear(10, 1)

  def forward(self, x):
    y=self.linear(x)
    y=torch.sigmoid(y)
    y=self.linear1(y)
    return(y)

```

#Model and Data

```
# Initialize the model
model = MyModel()
optimizer = optim.SGD(model.parameters(), lr=0.01)
#optimizer=optim.Adagrad(model.parameters(), lr=0.01)
# Generate random input and output
X = torch.randn(1000, 10)
Y_gt = torch.randn(1000, 1)
```

```
Y_gt
```

#Dig Down

```
# model  trainable parameters with name ans shape
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.shape,print(param))
```

```
X.shape
```

**Forward and Backward pass**

```
#Forward Pass
Y=model(X)

# Backward pass
loss=torch.mean(torch.abs(Y-Y_gt))
optimizer.zero_grad()
loss.backward()
optimizer.step()
```

```
Y=model(X)
print("input shape: ",X.shape)
print("ouput Shape",Y.shape)
print(torch.max(Y))
```

#Training Loop#

```
for  e_idx in range(1000):
  Y=model(X)

  loss=torch.mean(torch.abs(Y-Y_gt))

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  # model  trainable parameters with name ans shape
  for name, param in model.named_parameters():
      if param.requires_grad:
          print(name, param.shape,print(param))


  if  e_idx%10==0:
    print(f"epoch {e_idx} loss {loss.item()}")
```
