```
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
import  torch.nn.functional  as F

# Define the generator network
class Generator(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Generator, self).__init__()
        self.fc1 = nn.Linear(input_dim, 256)
        self.fc2 = nn.Linear(256, 512)
        self.fc3 = nn.Linear(512, output_dim)

    def forward(self, x):
        x = F.leaky_relu(self.fc1(x), 0.2)
        x = F.leaky_relu(self.fc2(x), 0.2)
        x = self.fc3(x)
        return x

# Define the discriminator network
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(input_dim, 256)
        self.fc2 = nn.Linear(256, 1)

    def forward(self, x):
        x = F.leaky_relu(self.fc1(x), 0.2)
        x = self.fc2(x)
        return torch.sigmoid(x)

# Define the loss function
loss = nn.BCELoss()

# Define the training data
# Replace this with your own data
data = np.random.normal(5, 1, size=(100, 10))

# Convert the training data to a PyTorch tensor
# This is the fix to convert the NumPy array to a PyTorch tensor
data = torch.tensor(data, dtype=torch.float32)

# Train the GAN
generator = Generator(10, 10)
discriminator = Discriminator(10)
generator_optim = optim.Adam(generator.parameters())
discriminator_optim = optim.Adam(discriminator.parameters())

for epoch in range(100):
    # Train the generator
    generator_optim.zero_grad()
    noise = torch.randn(100, 10)
    fake_data = generator(noise)
    # Fix: Create a target tensor of shape [100, 1]
    generator_loss = loss(discriminator(fake_data), torch.ones((100, 1)))
    generator_loss.backward()
    generator_optim.step()

    # Train the discriminator
    discriminator_optim.zero_grad()
    # Convert the real data to a PyTorch tensor for each batch
    real_data_tensor = torch.tensor(data, dtype=torch.float32)
    real_loss = loss(discriminator(real_data_tensor), torch.ones_like(discriminator(real_data_tensor)))
    fake_loss = loss(discriminator(fake_data.detach()), torch.zeros_like(discriminator(fake_data.detach()))) # Detach fake_data to avoid double backprop
    discriminator_loss = (real_loss + fake_loss) / 2
    discriminator_loss.backward()
    discriminator_optim.step()

    # Print the losses
    print(f"Epoch: {epoch + 1}, Generator loss: {generator_loss:.4f}, Discriminator loss: {discriminator_loss:.4f}")

    if(epoch % 10 == 0):
      # Plot the real and fake data
      plt.hist(data.numpy().flatten(), bins=100, label="Real data")
      plt.hist(fake_data.detach().numpy().flatten(), bins=100, label="Fake data")
      plt.legend()
      plt.show()



# Generate some fake data
fake_data = generator(noise)

# Plot the real and fake data
plt.hist(data.numpy().flatten(), bins=100, label="Real data")
plt.hist(fake_data.detach().numpy().flatten(), bins=100, label="Fake data")
plt.legend()
plt.show()
```
